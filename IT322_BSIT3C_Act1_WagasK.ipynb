## Importing Pandas and Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Task 1: Load the Dataset

```python
df = pd.read_csv("SalesDataCSV.csv")
df.head()
```

### View of Data Sales Display

```python
df
```

## Task 2: Explore and Clean the Dataset

```python
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
df = df.dropna(subset=["Product"]) 
df.loc[3, "Price"] = 700 
df.loc[5, "Price"] = 50

df["Total"] = df["Total"].fillna(df["Quantity"] * df["Price"])

df
```

## Task 3: Data Analysis and Manipulation

```python
top_5_products = df.sort_values(by="Total", ascending=False).head(5)
top_5_products
```

## Task 4: Data Visualization

```python

product_sales = df.groupby("Product")["Total"].sum().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
product_sales.plot(kind="bar", color="green", edgecolor="black")

plt.title("Total Sales per Product", fontsize=14)
plt.xlabel("Product", fontsize=12)
plt.ylabel("Total Sales ($)", fontsize=12)
plt.xticks(rotation=45, ha="right")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

### Line Chart (Monthly Sales)

```python
df["Month"] = df["Date"].dt.to_period("M")
monthly_sales = df.groupby("Month")["Total"].sum()

plt.figure(figsize=(10, 5))
monthly_sales.plot(marker="o", linestyle="-", color="red")

plt.title("Monthly Sales Trend", fontsize=14)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Total Sales ($)", fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, linestyle="--", alpha=0.7)
plt.show()
```

# Sales Data Analysis Report

## 1. Introduction
This report summarizes the analysis of sales data, covering data exploration, cleaning, and visualization of key insights. The primary objectives were to:  

·Clean and preprocess the dataset.  

·Analyze total sales per product and category.  

·Identify the top 5 best-selling products.

## 2. Data Exploration and Cleaning

### Dataset Overview

The dataset consists of 30 sales transactions, including attributes such as order ID, date, product name, category, quantity, price, and total sales amount.  

Handling Missing Values  

·Missing product names were replaced with "Unspecified."  

·Missing quantities were set to 0.  

·Missing prices were filled with the median price.  

·Missing total sales values were computed as Quantity * Price when applicable.  


## 3. Data Analysis

### Total Sales per Category

Total sales were aggregated based on product categories to determine which category generated the most revenue.  

### Top 5 Best-Selling Products

The dataset was grouped by product name, and total sales were calculated. The top 5 highest-revenue products were identified.  


## 4. Data Visualization

4.1 Total Sales by Category  

A bar chart was created to illustrate the total sales revenue generated by each category.  

4.2 Top 5 Best-Selling Products  

A bar chart was created to highlight the highest-selling products based on total revenue.  


## 5. Key Insights and Visualization

### Sales by Category:

·The highest sales come from the Electronics category.  

·Furniture and Clothing follow but with significantly lower total sales.  


```python
category_sales = df.groupby("Category")["Total"].sum().sort_values(ascending=False)

plt.figure(figsize=(8, 5))
category_sales.plot(kind="barh", color="purple", edgecolor="black")  

plt.xlabel("Total Sales ($)", fontsize=12)
plt.ylabel("Category", fontsize=12)
plt.title("Total Sales by Category", fontsize=14)
plt.grid(axis="x", linestyle="--", alpha=0.7)  
plt.show()
```

### Best-Selling Products:

·The Laptop and Smartphone lead in revenue, confirming the dominance of the Electronics category.  

·Other top-selling products include Sofas and T-Shirts but at a lower scale.  

```python
top_products = df.groupby("Product")["Total"].sum().sort_values(ascending=False).head(5)

plt.figure(figsize=(8, 5))
top_products.plot(kind="barh", color="orange", edgecolor="black")  

plt.xlabel("Total Sales ($)", fontsize=12)
plt.ylabel("Product", fontsize=12)
plt.title("Top 5 Best-Selling Products", fontsize=14)
plt.grid(axis="x", linestyle="--", alpha=0.7)  
plt.show()
```
